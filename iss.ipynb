```python
import requests
import time
import matplotlib.pyplot as plt
from datetime import datetime
```

```python
latitudes = []
longitudes = []
timestamps = []
```

```python
url = "https://api.wheretheiss.at/v1/satellites/25544"

for i in range(100):
    response = requests.get(url)
    
    if response.status_code == 200:
        data = response.json()

        lat = data['latitude']
        lon = data['longitude']
        ts = datetime.utcfromtimestamp(data['timestamp']).strftime('%Y-%m-%d %H:%M:%S')

        latitudes.append(lat)
        longitudes.append(lon)
        timestamps.append(ts)

        print(f"{i+1:03}: {ts} | Latitude: {lat:.2f}, Longitude: {lon:.2f}")
    else:
        print(f"Failed at point {i+1}")
    
    time.sleep(10)
```

```python
plt.figure(figsize=(12, 6))
plt.plot(longitudes, latitudes, 'bo-', markersize=3)
plt.title("ISS Path (Latitude vs Longitude)")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.grid(True)
plt.tight_layout()
plt.show()
```
